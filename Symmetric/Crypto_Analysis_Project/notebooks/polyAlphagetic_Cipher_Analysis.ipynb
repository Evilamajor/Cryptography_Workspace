```python
# Definici贸 expl铆cita dels textos
original = "Eduard encriptacio"
encrypted = "Jnzuwm aspvwudfpnx"
pattern = ['Caesar', 'Monoalphabetic']  # Alternan莽a clarament especificada

# Construir taula expl铆cita
data = []
pattern_applied = []

j = 0  # Comptador nom茅s per carcters alfab猫tics
for i, char in enumerate(original):
    if char.isalpha():
        cipher_type = pattern[j % len(pattern)]
        data.append([char, encrypted[i], cipher_type])
        pattern_applied.append(cipher_type)
        j += 1
    else:
        data.append([char, encrypted[i], 'Cap (espai)'])
        pattern_applied.append('Cap (espai)')

# DataFrame clar i visual
df = pd.DataFrame(data, columns=['Carcter Original', 'Carcter Xifrat', 'Tipus de Xifrat'])
df
```

```python
# Eliminar espais i signes de puntuaci贸
original_clean = ''.join(char for char in original if char.isalpha())
encrypted_clean = ''.join(char for char in encrypted if char.isalpha())

# Mostrar resultats clarament
print("Missatge original sense espais ni puntuaci贸:", original_clean)
print("Missatge xifrat sense espais ni puntuaci贸:", encrypted_clean)
```
